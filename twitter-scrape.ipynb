```python
import nest_asyncio
import twint
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

import re
from nltk.tokenize import WordPunctTokenizer
from nltk.corpus import stopwords

from textblob import TextBlob

from wordcloud import WordCloud
```

```python
nest_asyncio.apply()

keywords = {"astrotraffic":"kltrafficupdate"}
#"Malaysia":"Motor insurance OR \"vehicle insurance\" OR \"car insurance\"",
#"Malaysia":"flood insurance",
#"Malaysia flood": "#malaysiaflood"}

def twintConfig(date_from,date_to,search_string):
    c = twint.Config()
    c.Search = search_string[1]
    c.Since = date_from
    c.Until = date_to
    c.Pandas = True
    twint.run.Search(c)

since = "2021-09-08"#input ("Start date: ")
until = "2022-09-09"#input ("End date: ")
    
    
def Run_Twint(search_values):
    df = pd.DataFrame()
    
    for keyword in search_values.items():
        twintConfig(since,until,keyword)
        tweets_df = twint.storage.panda.Tweets_df
        tweets_df["Keyword"] = keyword[0]
        df = pd.concat([df,tweets_df])
    
    return df
```

```python
tweets_df = Run_Twint(keywords)
```
